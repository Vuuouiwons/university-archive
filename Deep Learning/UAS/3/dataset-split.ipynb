# Train test split

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('./dataset/preprocessed-dataset.csv')
df_bin = pd.read_csv('./dataset/preprocessed-dataset-binary.csv')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['clean_text'], df['label'], test_size=0.3, stratify=df['label'])
x_train_bin, x_test_bin, y_train_bin, y_test_bin = train_test_split(df_bin['clean_text'], df_bin['label'], test_size=0.3, stratify=df_bin['label'])
```

```python
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.5, stratify=y_test)
x_val_bin, x_test_bin, y_val_bin, y_test_bin = train_test_split(x_test_bin, y_test_bin, test_size=0.5, stratify=y_test_bin)
```

```python
# --- Recombine into DataFrames ---
train_set = pd.concat([x_train, y_train], axis=1).reset_index(drop=True)
val_set = pd.concat([x_val, y_val], axis=1).reset_index(drop=True)
test_set = pd.concat([x_test, y_test], axis=1).reset_index(drop=True)
train_set_bin = pd.concat([x_train_bin, y_train_bin], axis=1).reset_index(drop=True)
val_set_bin = pd.concat([x_val_bin, y_val_bin], axis=1).reset_index(drop=True)
test_set_bin = pd.concat([x_test_bin, y_test_bin], axis=1).reset_index(drop=True)

# # (Optional) Rename columns for consistency
# train_set.columns = val_set.columns = test_set.columns = df.columns
# train_set_bin.columns = val_set_bin.columns = test_set_bin.columns = df_bin.columns
```

```python
train_set.to_csv('./dataset/dataset_train_set.csv')
val_set.to_csv('./dataset/dataset_val_set.csv')
test_set.to_csv('./dataset/dataset_test_set.csv')
train_set_bin.to_csv('./dataset/dataset_train_set_bin.csv')
val_set_bin.to_csv('./dataset/dataset_val_set_bin.csv')
test_set_bin.to_csv('./dataset/dataset_test_set_bin.csv')
```
