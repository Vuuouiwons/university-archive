```python
PUNCTUATION = set(string.punctuation + "“”‘’´`′‵–—−…•·∙«»※⁂¤§¶")

def normalize_text(text: str) -> str:
    """Normalize unicode and lowercase text."""
    return unicodedata.normalize("NFKC", str(text)).lower().strip()

def remove_punctuation(words: list[str]) -> list[str]:
    """Remove punctuation characters from within tokens."""
    parsed_words = []
    for word in words:
        cleaned = ''.join(ch for ch in word if ch not in PUNCTUATION)
        if cleaned.strip():
            parsed_words.append(cleaned)
    return parsed_words

def remove_numbers(words: list[str]) -> list[str]:
    """Remove numeric tokens or tokens containing digits."""
    parsed_words = []
    for word in words:
        if re.search(r'\d', word):  # remove if contains any digit
            continue
        parsed_words.append(word)
    return parsed_words

def remove_stop_words(words: list[str]) -> list[str]:
    """Remove stopwords."""
    return [w for w in words if w not in STOP_WORDS]

# ======================
# MAIN PIPELINE
# ======================
df_binary = pd.read_csv(raw_dataset, index_col='index')

df_binary['text'] = df_binary['text'].apply(normalize_text)
df_binary['tokenized'] = df_binary['text'].apply(lambda x: nltk.word_tokenize(str(x)))
df_binary['tokenized_punct'] = df_binary['tokenized'].apply(remove_punctuation)
df_binary['tokenized_no_num'] = df_binary['tokenized_punct'].apply(remove_numbers)
df_binary['tokenized_clean'] = df_binary['tokenized_no_num'].apply(remove_stop_words)
df_binary['clean_text'] = df_binary['tokenized_clean'].apply(lambda x: ' '.join(x))

df_binary.head()
```
