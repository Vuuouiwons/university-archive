```python
from bs4 import BeautifulSoup as bs
```

```python
url_detik = 'https://sport.detik.com/sepakbola/liga-indonesia/d-8203344/piala-dunia-u-17-indonesia-sudah-kantongi-kelebihan-honduras'
url_liputan6 = 'https://www.liputan6.com/bola/read/6208059/scott-mctominay-dorong-napoli-gaet-kobbie-mainoo-dari-mu?page=4'
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
options.add_argument("--headless=new")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
driver.get(url_detik)
time.sleep(3)

html = driver.page_source
soup = bs(html, 'html.parser')

for p in soup.find_all('p'):
    p = p.get_text(strip=True)
    print(p)
```

```python
driver.get(url_liputan6)
time.sleep(3)

html = driver.page_source
soup = bs(html, 'html.parser')
```

```python
for p in soup.select('.read-page-upper .article-content-body__item-content'):
    p = p.get_text(strip=True)
    print(p, sep='\n')
```
