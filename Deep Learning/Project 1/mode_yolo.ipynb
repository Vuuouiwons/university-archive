```python
from ultralytics import YOLO
```

```python
# Load pretrain model
model = YOLO("yolo11n-cls.yaml").load("yolo11n-cls.pt")

model.info(detailed=True)
```

# Train model

```python
results = model.train(cfg="./yolo_classify/yolo11n-cls.yaml")
```

```python
model.info(detailed=True)
```

Framework automatically droped existing output layer which contains 1000 class and replaced it with 7 classes aligned with the FER2013 dataset

```python
metrics = model.val(data="dataset.yaml", split="test", save=True)

print(metrics)
```
