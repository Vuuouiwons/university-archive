```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
dataset_path = os.path.join(os.getcwd(), 'dataset')
embedded_dataset = os.path.join(dataset_path, 'embedded_spam_message_original.pkl')
test_embedded_dataset = os.path.join(dataset_path, 'embedded_spam_message_test.pkl')
```

```python
df = pd.read_pickle(embedded_dataset)
```

```python
df.head()
```

```python
X = df['embed_vector_padded']
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, stratify=y)

(X_train, y_train)
```

```python
train_df = pd.DataFrame({
    'embed_vector_padded': X_train,  # Or 'embeddings', or your original column name for X
    'class': y_train      # Or your original column name for y
})

test_df = pd.DataFrame({
    'embed_vector_padded': X_test,   # Use the same column name as above
    'class': y_test       # Use the same column name as above
})
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
test_df = test_df.dropna()
test_df.head()
```

```python
test_df.to_pickle(test_embedded_dataset)
```
